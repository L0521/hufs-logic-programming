```python
#인공신경망을 통해서 60가지 성질을 통해 암석과 광물을 구분
#202302834 이승준

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt


targetUrl = "https://raw.githubusercontent.com/gundaminpde/2022/main/sonar3.csv"


Data_set = pd.read_csv(targetUrl, sep=',')

# 입력받는 X는 특성, y는 정답
X = Data_set.iloc[:,0:60]
y = Data_set.iloc[:,60]


print(Data_set.shape)
Data_set.head()

model = Sequential()
model.add(Dense(30, input_dim=60, activation='relu'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['binary_accuracy'])
history=model.fit(X, y, epochs=60, batch_size=20)


y_loss = history.history['loss']

x_len = np.arange(len(y_loss))

plt.plot(x_len,y_loss)
```
